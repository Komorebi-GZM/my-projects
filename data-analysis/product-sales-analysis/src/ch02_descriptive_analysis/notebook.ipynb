# ch02 描述性统计与可视化

本 Notebook 对清洗后的产品销售数据进行描述性统计与可视化分析，包括描述性统计、类别分析、城市分析和时间趋势分析。

## 1. 环境配置与数据加载

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path('.').resolve()
sys.path.insert(0, str(PROJECT_ROOT))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from src.utils import (
    load_preprocessed,
    save_dataframe,
    OUTPUT_BASE,
    CATEGORY_LIST,
    CITY_LIST,
)
from src.utils.visualizer import (
    plot_category_sales,
    plot_city_sales,
    plot_time_series,
    plot_heatmap,
    plot_category_distribution,
)
from src.utils.metrics import (
    calc_total_sales,
    calc_avg_price,
    calc_category_summary,
    calc_city_summary,
)

print('环境配置完成。')
```

```python
# 加载清洗后的数据
df = load_preprocessed('ch01_data_cleaning', 'product')
print(f'数据形状: {df.shape}')
print(f'数据列: {df.columns.tolist()}')
df.head()
```

## 2. 描述性统计分析

```python
# 描述性统计
desc_stats = df.describe(include='all')
print('描述性统计:')
desc_stats
```

```python
# 关键指标
total_sales = calc_total_sales(df)
avg_price = calc_avg_price(df)
print(f'数据总量: {df.shape[0]} 条记录')
```

## 3. 产品类别分析

```python
# 类别汇总统计
category_summary = calc_category_summary(df)
print('类别汇总:')
category_summary
```

```python
# 类别销售额柱状图
output_dir = OUTPUT_BASE / 'descriptive_analysis'
output_dir.mkdir(parents=True, exist_ok=True)
plot_category_sales(df, save_path=output_dir / 'category_sales.png')
plt.show()
```

```python
# 类别分布饼图
plot_category_distribution(df, save_path=output_dir / 'category_distribution.png')
plt.show()
```

## 4. 城市维度分析

```python
# 城市汇总统计
city_summary = calc_city_summary(df)
print('城市汇总:')
city_summary
```

```python
# 城市销售额对比图
plot_city_sales(df, save_path=output_dir / 'city_sales.png')
plt.show()
```

## 5. 时间趋势分析

```python
# 销售额时间趋势
if 'date' in df.columns:
    plot_time_series(df, save_path=output_dir / 'sales_time_series.png')
    plt.show()
else:
    print('数据中无日期列，跳过时间趋势分析。')
```

```python
# 月度销售额趋势
if 'date' in df.columns:
    df_copy = df.copy()
    df_copy['date'] = pd.to_datetime(df_copy['date'])
    df_copy['month'] = df_copy['date'].dt.to_period('M')
    monthly_sales = df_copy.groupby('month')['total_price'].sum()
    print('月度销售额趋势:')
    print(monthly_sales)
```

## 6. 相关性分析

```python
# 相关性热力图
plot_heatmap(df, save_path=output_dir / 'correlation_heatmap.png')
plt.show()
```

## 7. 分析总结

```python
print('=' * 50)
print('描述性统计与可视化 - 分析总结')
print('=' * 50)
print(f'总销售额: {total_sales:,.2f}')
print(f'平均价格: {avg_price:,.2f}')
print(f'数据总量: {df.shape[0]} 条记录')
if len(category_summary) > 0:
    print(f'销售额最高类别: {category_summary.iloc[0]["category"]}')
if len(city_summary) > 0:
    print(f'销售额最高城市: {city_summary.iloc[0]["city"]}')
print('=' * 50)
print('分析结果已保存至 outputs/descriptive_analysis/')
```
